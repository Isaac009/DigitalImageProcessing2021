```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('farao.jpg')
original = img.copy()
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
edge = cv2.Canny(img,100,200)
contours, hierarchy = cv2.findContours(edge,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
con = cv2.drawContours(img, contours, -1, (0,255,0), 3)
```

```python
img1 = cv2.imread('mikono.jpg')
original1 = img1.copy()
imgray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
ret1,thresh1 = cv2.threshold(imgray1,50,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
edge1 = cv2.Canny(imgray1,100,200)
contours1, hierarchy1 = cv2.findContours(edge1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.imread('nike.jpg')
# ret,thresh = cv2.threshold(im,127,255,0)
```

```python
ret,th=cv2.threshold(im,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours,hierarchy = cv2.findContours(th, cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
cnt = contours[1]
M = cv2.moments(cnt)
```

```python
im.shape
```

```python
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
```

```python
img = np.zeros(im.shape, np.uint8)
img = cv2.circle(img,(cx,cy),3,(0,255,0),2)
```

```python
img = cv2.drawContours(img, contours[1], -1, (0,255,0), 3)
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(121),plt.imshow(im),plt.title('Original', fontsize=10)
plt.subplot(122),plt.imshow(img, 'gray'),plt.title('Contour Image', fontsize=10)
```
