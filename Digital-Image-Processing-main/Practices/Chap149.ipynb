```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('farao.jpg')
original = img.copy()
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
edge = cv2.Canny(img,100,200)
contours, hierarchy = cv2.findContours(edge,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
con = cv2.drawContours(img, contours, -1, (0,255,0), 3)
```

```python
img1 = cv2.imread('mikono.jpg')
original1 = img1.copy()
imgray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
ret1,thresh1 = cv2.threshold(imgray1,50,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
edge1 = cv2.Canny(imgray1,100,200)
contours1, hierarchy1 = cv2.findContours(edge1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cont1 = cv2.drawContours(img1, contours1, -1, (255, 0,0), 3)
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(231),plt.imshow(original),plt.title('Original', fontsize=10)
plt.subplot(232),plt.imshow(edge, 'gray'),plt.title('Canny Edge', fontsize=10)
plt.subplot(233),plt.imshow(con),plt.title('Countor', fontsize=10)
plt.subplot(234),plt.imshow(original1),plt.title('Original', fontsize=10)
plt.subplot(235),plt.imshow(edge1, 'gray'),plt.title('Canny Edge', fontsize=10)
plt.subplot(236),plt.imshow(cont1),plt.title('Countor', fontsize=10)
plt.show()
```
