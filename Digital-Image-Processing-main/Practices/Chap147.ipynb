```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('tnj.jpg',0)
blur = cv2.GaussianBlur(img,(5,5),0)
edges = cv2.Canny(img,100,200)
blur_edge = cv2.Canny(blur, 100,125)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(221),plt.imshow(img, 'gray'),plt.title('Original', fontsize=20)
plt.subplot(222),plt.imshow(blur, 'gray'),plt.title('Blur', fontsize=20)
plt.subplot(223),plt.imshow(edges, 'gray'),plt.title('Canny Edges', fontsize=20)
plt.subplot(224),plt.imshow(blur_edge, 'gray'),plt.title('Blur + Canny', fontsize=20)
plt.show()
```

```python
img = cv2.imread('tomjerry.jpg')
down = cv2.pyrDown(img)
down1 = cv2.pyrDown(down)
down2 = cv2.pyrDown(down1)
down3 = cv2.pyrDown(down2)
up = cv2.pyrUp(down3)
up1 = cv2.pyrUp(up)
up2 = cv2.pyrUp(up1)
up3 = cv2.pyrUp(up2)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(241),plt.imshow(img),plt.title('Original', fontsize=10)
plt.subplot(242),plt.imshow(down),plt.title('Down 1', fontsize=10)
plt.subplot(243),plt.imshow(down1),plt.title('Down 2', fontsize=10)
plt.subplot(244),plt.imshow(down2),plt.title('Down 3', fontsize=10)
plt.subplot(245),plt.imshow(up),plt.title('Up 1', fontsize=10)
plt.subplot(246),plt.imshow(up1),plt.title('Up 2', fontsize=10)
plt.subplot(247),plt.imshow(up2),plt.title('Up 3', fontsize=10)
plt.subplot(248),plt.imshow(up3),plt.title('Up 4', fontsize=10)
plt.show()
```
