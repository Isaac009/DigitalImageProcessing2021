```python
## 1.4.5 Morphological Transformations
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('character.jpeg',0)
kernel = np.ones((2,2),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(231),plt.imshow(img, 'gray'),plt.title('Input', fontsize=20)
plt.subplot(232),plt.imshow(erosion, 'gray'),plt.title('Output Erosion', fontsize=20)
plt.subplot(233),plt.imshow(dilation, 'gray'),plt.title('Output: Dilation', fontsize=20)
plt.subplot(234),plt.imshow(opening, 'gray'),plt.title('Output: Opening', fontsize=20)
plt.subplot(235),plt.imshow(closing, 'gray'),plt.title('Output: Closing', fontsize=20)
plt.subplot(236),plt.imshow(gradient, 'gray'),plt.title('Output: Gradient', fontsize=20)
plt.show()
```

```python
## Enhance Fingerprint using mophological operators 
```

```python
img = cv2.imread('finger1.jpeg',0)
kernel = np.ones((2,2),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations = 2)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(231),plt.imshow(img, 'gray'),plt.title('Input', fontsize=20)
plt.subplot(232),plt.imshow(erosion, 'gray'),plt.title('Output Erosion', fontsize=20)
plt.subplot(233),plt.imshow(dilation, 'gray'),plt.title('Output: Dilation', fontsize=20)
plt.subplot(234),plt.imshow(opening, 'gray'),plt.title('Output: Opening', fontsize=20)
plt.subplot(235),plt.imshow(closing, 'gray'),plt.title('Output: Closing', fontsize=20)
plt.subplot(236),plt.imshow(gradient, 'gray'),plt.title('Output: Gradient', fontsize=20)
plt.show()
```
